```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
datasets_train = pd.read_csv("./input/train.csv")
datasets_test = pd.read_csv("./input/test.csv")
```

```python
datasets_train.head()
```

```python
datasets_test.head()
```

```python
def get_X_y(datasets):
    return datasets_train.drop("label", axis=1), datasets_train["label"].copy()

X_train, y_train = get_X_y(datasets_train)
X_test = datasets_test
```

```python
X_train.head()
```

```python
X_train.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

rf_clf = RandomForestClassifier()
score = cross_val_score(
    rf_clf, 
    X_train, 
    y_train, 
    cv=3, 
    scoring="accuracy", 
    n_jobs=-1,
    verbose=2
)
score.mean()
```

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

rf_clf = RandomForestClassifier()
param_distribs = {
    "max_features": randint(low=78, high=784),
#     "n_estimators": randint(low=10, high=100),
}

rnd_search = RandomizedSearchCV(
    rf_clf,
    param_distributions=param_distribs,
    n_iter=10,
    cv=3,
    scoring="accuracy",
    verbose=2,
    n_jobs=-1,
)
rnd_search.fit(X_train, y_train)
print(rnd_search.best_params_)
print(rnd_search.best_score_)
```

```python
final_model1 = rnd_search.best_estimator_
```

```python
rf_clf = RandomForestClassifier()
param_distribs = {
#     "max_features": randint(low=78, high=784),
    "n_estimators": randint(low=100, high=200),
}

rnd_search = RandomizedSearchCV(
    rf_clf,
    param_distributions=param_distribs,
    n_iter=10,
    cv=3,
    scoring="accuracy",
    verbose=2,
    n_jobs=-1,
)
rnd_search.fit(X_train, y_train)
print(rnd_search.best_params_)
print(rnd_search.best_score_)
final_model2 = rnd_search.best_estimator_
```

```python
rf_clf = RandomForestClassifier()
param_distribs = {
#     "max_features": randint(low=78, high=784),
    "n_estimators": randint(low=500, high=1000),
}

rnd_search = RandomizedSearchCV(
    rf_clf,
    param_distributions=param_distribs,
    n_iter=10,
    cv=3,
    scoring="accuracy",
    verbose=2,
    n_jobs=-1,
)
rnd_search.fit(X_train, y_train)
print(rnd_search.best_params_)
print(rnd_search.best_score_)
final_model3 = rnd_search.best_estimator_
```

```python

def save_submission(model, X_test, try_times=0):
    y_pred = model.predict(X_test)
    data = np.c_[np.array([idx + 1 for idx in range(len(y_pred))]), y_pred]
    y_pred_csv = pd.DataFrame(data=data, columns=["ImageId", "Label"])
    y_pred_csv.to_csv("./output/submission_%(try_times)02d.csv" % {"try_times": try_times}, index=False)
        
```

```python
save_submission(final_model2, X_test, 1)
```
